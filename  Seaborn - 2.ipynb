```python
import seaborn as sns
```

```python
dataset = sns.load_dataset('tips')
```

```python
dataset.head()
```

```python
dataset.dtypes
```

# Correlation with Heatmap

A correaltion heatmap uses colored cells, typically in monochromatic scale, to show a 2D correlation matrix between two discrete dimensions or event types.

```python
dataset.corr()
```

```python
sns.heatmap(dataset.corr())
```

## Joinplot

A join plot allows the relationship between two numeric variables. The central chart disply their correlation. It is usually a scatterplot, a hixbin plot, 1 2D histogrm or a 2D density plot.


```python
sns.jointplot(x = 'tip', y= 'total_bill', data = dataset, kind = 'hex')
```

```python
sns.jointplot(x = 'tip', y= 'total_bill', data = dataset, kind = 'reg')
```

### Pair Plot
A pair plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaboratins on this , showing all variables paired with all the other variables. 

```python
sns.pairplot(data = dataset)
```

```python
sns.pairplot(data = dataset, hue = 'sex')
```

```python
sns.pairplot(data = dataset, hue = 'smoker', kind = 'reg')
```

```python
sns.pairplot(data = dataset, hue = 'day', diag_kind = 'kde')
```

### Dist Plot
Dist plot helps us to check the distribution of the columns features.

```python
sns.distplot(dataset['tip'])
```

```python
sns.distplot(dataset['total_bill'], kde = False, bins = 10)
```

# Categorical Plots
Seaborn also helps us in doing the analysis on categorical Data Points. 
(a) Count Plot 
(b) Bar Plot
(c) Box Plot
(d) Violin Plot

```python
sns.countplot(dataset['sex']) # countplot('sex', data = dataset)
```

```python
sns.countplot(y = 'day', data = dataset, palette = 'rainbow')
```

```python
sns.countplot(y = 'smoker', data = dataset)
```

### Bar Plot
A bar plot represents an estimate of central tendency for a numeric
variable with the height of each rectangle and provides some indication of
the uncertainty around that estimate using error bars. 

```python
sns.barplot(x = 'total_bill', y = 'smoker', data = dataset)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = dataset)
```

### Box Plot
A box and wisker plot (also boxplot) is a graph that presents information from a five-number summary. 

```python
sns.boxplot(x = 'total_bill', y = 'smoker', data = dataset) # the polygon is called OUTLIERs
```

```python
sns.boxplot(x = 'tip', y = 'day', data = dataset, palette = 'rocket')
```

```python
sns.boxplot(data = dataset, orient = 'v', palette = 'cubehelix')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', hue = 'smoker', data = dataset, palette = 'viridis')
```

### Violin Plot
Violin plot helps us to see both the distribution of data in items of Kernal density estimation and the box plot.

```python
sns.violinplot(x = 'total_bill', y = 'day', data = dataset, palette = 'rainbow')
```
